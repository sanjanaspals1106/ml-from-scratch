# Linear Regression — Geometric View

A linear model predicts outputs by combining inputs linearly.
Geometrically, this means choosing a direction (weights) and a shift (bias)
that best aligns with the data.

The model does not “solve equations” — it **fits a line (or plane) in space**
so that predictions are as close as possible to the actual points.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x=np.linspace(0,10,50)
true_w=2.5
true_b=1.0
noise=np.random.normal(0,2,size=x.shape)
y=true_w*x+true_b+noise
```

```python
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Synthetic data")
plt.show()
```

```python
def predict(x,w,b):
    return w*x+b
```

```python
w_bad=0.5
b_bad=5.0
```

```python
y_bad = predict(x, w_bad, b_bad)
```

```python
plt.scatter(x, y)
plt.plot(x, y_bad)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Underfitting example")
plt.show()
```

```python
w_good=2.3
b_good=1.2
```

```python
y_good=predict(x, w_good, b_good)
```

```python
plt.scatter(x,y)
plt.plot(x, y_good)
plt.title("Better fit")
plt.show()
```

```python
residuals=y-y_good
```

```python
plt.scatter(x,y)
plt.plot(x, y_good)
for xi, yi, ypi in zip(x, y, y_good):
    plt.plot([xi,xi],[yi,ypi])
plt.title("Residuals Visualization")
plt.show()
```
