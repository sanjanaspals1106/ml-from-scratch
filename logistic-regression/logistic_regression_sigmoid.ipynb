# Logistic Regression — Sigmoid Function

In binary classification, the target variable takes only two values:
y ∈ {0, 1}.

A linear model produces a real-valued score:
z = wᵀx + b

This score indicates **confidence**, but it is not a probability because it is
unbounded and cannot be interpreted directly.

To convert this confidence score into a meaningful probability, we apply
the sigmoid function:

σ(z) = 1 / (1 + e⁻ᶻ)

The sigmoid maps any real number to the interval (0, 1), allowing the model
to output:

P(y = 1 | x) = σ(wᵀx + b)

This notebook focuses only on **why** sigmoid is needed and how it converts
linear scores into probabilities. Training and loss functions are covered later.


```python
import numpy as np

def sigmoid(z):
    np.clip(z,-500,500)
    return 1/(1+np.exp(-z))
```

```python
def predict_prob(x, w, b):
    z=x@w + b
    return sigmoid(z)
```

```python
import matplotlib.pyplot as plt

z = np.linspace(-10, 10, 200)
s = sigmoid(z)

plt.plot(z, s)
plt.axhline(0.5)
plt.axvline(0)
plt.xlabel("z = w·x + b")
plt.ylabel("σ(z)")
plt.title("Sigmoid Function")
plt.show()
```

## Key Takeaway

Logistic regression does not change the linear nature of the model.

It first computes a linear confidence score:
z = wᵀx + b

The sigmoid function then **calibrates** this score into a probability:

P(y = 1 | x) = σ(z)

This allows the model to make probabilistic predictions while retaining a
linear decision boundary.

Sigmoid is needed not to "fit a curve", but to convert unbounded confidence
into interpretable probability.
